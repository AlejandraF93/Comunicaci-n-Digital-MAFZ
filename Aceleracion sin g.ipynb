```python
#Llamado de librerias
import pandas as pd
import matplotlib.pyplot as plt

#Cargar el archivo .CSV
file_path = "Exp1.csv"
data=pd.read_csv(file_path, delimiter='\t')
data.columns = data.columns.str.strip()
data = data.apply(lambda x: x.str.strip() if isinstance(x,str) else x)

#Renombrar columnas
Time_s_='Time (s)'
LinearAccelerationX_m_s_2_='Linear Acceleration x (m/s^2)'
LinearAccelerationY_m_s_2_='Linear Acceleration y (m/s^2)'
LinearAccelerationZ_m_s_2_='Linear Acceleration z (m/s^2)'

#Convertir columnas a numericos pd
data[Time_s_]=pd.to_numeric(data[Time_s_])
data[LinearAccelerationX_m_s_2_]=pd.to_numeric(data[LinearAccelerationX_m_s_2_])
data[LinearAccelerationY_m_s_2_]=pd.to_numeric(data[LinearAccelerationY_m_s_2_])
data[LinearAccelerationZ_m_s_2_]=pd.to_numeric(data[LinearAccelerationZ_m_s_2_])

#Grafica de aceleracin en X vs t(s)
plt.subplot(3,1,1)
plt.plot(data[Time_s_],data[LinearAccelerationX_m_s_2_])
plt.title('Aceleracin en X vs t(s)')
plt.xlabel('t(s)')
plt.ylabel('a(m/s^2)')
plt.grid(True)

#Grafica de aceleracin en Y vs t(s)
plt.subplot(3,1,2)
plt.plot(data[Time_s_],data[LinearAccelerationY_m_s_2_])
plt.title('Aceleracin en Y vs t(s)')
plt.xlabel('t(s)')
plt.ylabel('a(m/s^2)')
plt.grid(True)

#Grafica de aceleracin en Z vs t(s)
plt.subplot(3,1,3)
plt.plot(data[Time_s_],data[LinearAccelerationZ_m_s_2_])
plt.title('Aceleracin en Z vs t(s)')
plt.xlabel('t(s)')
plt.ylabel('a(m/s^2)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
