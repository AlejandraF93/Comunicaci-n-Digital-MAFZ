```python
#Llamado de librerias
import pandas as pd
import matplotlib.pyplot as plt

#Cargar el archivo .CSV
file_path = "Raw Data.csv"
data=pd.read_csv(file_path, delimiter='\t')
data.columns = data.columns.str.strip()
data = data.apply(lambda x: x.str.strip() if isinstance(x,str) else x)

#Renombrar columnas
Time_s_='Time (s)'
Illuminance_lx_='Illuminance (lx)'

#Convertir columnas a numericos pd
data[Time_s_]=pd.to_numeric(data[Time_s_])
data[Illuminance_lx_]=pd.to_numeric(data[Illuminance_lx_])

#Grafica de Medición de Luz
plt.subplot(3,1,1)
plt.plot(data[Time_s_],data[Illuminance_lx_])
plt.title(' Medición de Luz')
plt.xlabel('t(s)')
plt.ylabel('Ev(lx)')
plt.grid(True)
```
